## Create sample scenario

```python
df = pd.read_csv('eval_scenarios/eval_goal_gen_GT_scenarios.csv')

sampled_df = df.groupby('Goal Type', group_keys=False).apply(lambda x: x.sample(frac=35/len(df)))
```

```python
sampled_df.columns
```

```python
sampled_df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
sampled_df.reset_index(drop=True, inplace=True)
```

```python
sampled_df.to_csv('eval_scenarios/eval_goal_gen_GT_scenarios_sample.csv')
```
